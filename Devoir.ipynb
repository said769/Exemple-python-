```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction et gradient
def f(x, y):
    return 2 * x**2 + y**2 - x * y - 3 * x - y + 4

def grad_f(x, y):
    return np.array([4 * x - y - 3, 2 * y - x - 1])

# Gradient Ã  Pas Fixe (GPF)
def gradient_pas_fixe(x0, y0, gamma, n_iter):
    x, y = x0, y0
    trajectory = [(x, y)]
    for _ in range(n_iter):
        grad = grad_f(x, y)
        x, y = x - gamma * grad[0], y - gamma * grad[1]
        trajectory.append((x, y))
    return trajectory
```

```python

# Initialisation
x0, y0 = 0, 0
n_iter = 1

# Simulation
trajectory_gpf = gradient_pas_fixe(x0, y0, gamma=0.5, n_iter=n_iter)
```

```python
# Visualisation
x_vals_gpf, y_vals_gpf = zip(*trajectory_gpf)
```

```python
# Contour plot
x = np.linspace(-1, 2, 100)
y = np.linspace(-1, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.figure(figsize=(12, 6))
plt.contour(X, Y, Z, levels=50, cmap="viridis")
plt.plot(x_vals_gpf, y_vals_gpf, 'o-', label="GPF (gamma=0.5)", color="red")
```

```python
plt.scatter([1], [1], color="black", label="Minimum", zorder=5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Trajectoires des algorithmes de descente de gradient")
plt.legend()
plt.grid()
plt.show()
```
