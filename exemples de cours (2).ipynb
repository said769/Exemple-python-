Exemple 3:Resolution de l'équation ax2+bx+c=0 dans R

```python
from math import sqrt
a=float(input("donnez a:"))
b=float(input("donnez b:"))
c=float(input("donnez c:"))
D=b**2-4*a*c
if D<0 :
    print("pas solustion sur R:")
elif D==0 :
    x=-b/(2*a)
    print(x)
else :
    x1=-b-sqrt(D)/(2*a)
    x2=-b+sqrt(D)/(2*a)
    print(x1,x2)
print("fin du programme")
```

exemple 5:racine d'un entier positif

```python
## la boucle while
from math import sqrt
x=int(input("saisir un un entier positif:"))
while x<0 :
    x=int(input("saisir un un entier positif:"))
print ('la racine de ',x, 'est ',sqrt(x))
```
